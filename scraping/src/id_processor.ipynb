```python
import pandas as pd
```

```python
def process_id(frame):
    index = frame.index.to_series().str.strip()
    index = index.str.replace(" ", "-")
    frame.set_index(index, inplace=True)
    return frame
```

```python
caa = pd.read_excel("../../data/input/caa_courses.xlsx", index_col = "caa_course_id")
eq = pd.read_excel("../output/equivalencies2.xlsx", index_col = "caa_id")
eq = process_id(eq)
caa = process_id(caa)
known = pd.read_excel("../output/known.xlsx", index_col = "caa_id")
eq = eq.reset_index().drop_duplicates(subset='caa_id', keep='last').set_index('caa_id').sort_index()
eq
```
