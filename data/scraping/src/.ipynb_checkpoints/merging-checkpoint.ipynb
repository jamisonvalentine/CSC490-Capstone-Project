```python
import pandas as pd
```

```python
ss = pd.read_csv("ss_courses_sp2021.csv")
wa = pd.read_csv("wa_courses_sp2021.csv")
```

```python
ss.head()
```

```python
section_id = list(map(lambda x: x.split(" ")[0], list(ss['section'])))
course_id = list(map(lambda x: x[:7], section_id))
ss.insert(0, 'course_id', course_id)
ss.insert(1, 'section_id', section_id)
ss.set_index('section_id', inplace=True)
```

```python
section_id = list(map(lambda x: x.split(" ")[0], list(wa['section'])))
course_id = list(map(lambda x: x[:7], section_id))
wa.insert(0, 'course_id', course_id)
wa.insert(1, 'section_id', section_id)
wa.set_index('section_id', inplace=True)
```
